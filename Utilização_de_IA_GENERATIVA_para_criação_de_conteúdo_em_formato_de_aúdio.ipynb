#IMPORTAÇÕES

```python
import google.generativeai as genai
from gtts import gTTS
from IPython.display import Audio, display
from google.colab import userdata
```

#FUNÇÃO PARA GERAR CONTEÚDO

```python
def generate_content(topic, modelo):

  #Definições

  response = modelo.generate_content(topic)
  #print(f'Este artigo é um artigo sobre {topic}.\n \n O google Gemini ajudou a gerar este conteúdo: \n {response.text}')
  content = f'Este artigo é um artigo sobre {topic}.\n \n O google Gemini ajudou a gerar este conteúdo: \n {response.text}'
  return content
```

#FUNÇÃO PARA BUSCAR O MODELO

```python
def modelo(generation_config, safety_settings):
  API_KEY = userdata.get("API_KEY")
  genai.configure(api_key=API_KEY)
  model = genai.GenerativeModel("gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)
  return model
```

#FUNÇÃO PARA CONVERTER AUDIO

```python
def convert_to_audio(text, language='pt', filename='output.mp3', voice_name='pt-BR-Standard-C'):
    tts = gTTS(text=text, lang=language, slow=False)
    tts.save(filename)
    print(f"Áudio salvo como {filename}")
```

DEFINIÇÕES DE CONFIGURAÇÃO DE IA GENERATIVA

```python
generation_config = {
  "candidate_count": 1,
  "temperature": 0.5,
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}
```

#BLOCO PRINCIPAL

```python
topic = 'Impacto da inteligência artificial na medicina em no máximo 5 linhas'

model = modelo(generation_config, safety_settings)
text = generate_content(topic, model).replace("*", "")
convert_to_audio(text)
display(Audio('output.mp3', autoplay=True))
```
